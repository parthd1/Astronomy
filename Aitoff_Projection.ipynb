```python
import matplotlib.pyplot as plt
import csv
import numpy as np
```

```python
def main():
  tsv_file = open("Local_Group_Coords.tsv")
  read_tsv = csv.reader(tsv_file, delimiter="\t")
  next(read_tsv)
  #read_tsv=np.loadtxt("Local_Group_Coords.tsv")
  #RA=read_tsv[:, 1]
  #Dec=read_tsv[:,2]
  fig = plt.figure()
  fig.set_size_inches(20,40)
  ax = fig.add_subplot(111, projection= "hammer")
  plt.title("Local Group Hammer Aitoff Projection")
  plt.grid("True")

  for row in read_tsv:
    ax.scatter(x= float(row[1])*(np.pi/12), y= float(row[2])*(np.pi/180))
    ax.annotate(row[0], (float(row[1])*(np.pi/12), float(row[2])*(np.pi/180)))
  fig.show()
```

```python
main()
```
